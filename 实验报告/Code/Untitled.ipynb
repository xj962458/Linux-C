1、美团爬虫

```python
import requests
from bs4 import BeautifulSoup

r = requests.get('https://wuhu.meituan.com/', timeout=30)
r.encoding = 'utf-8'
soup = BeautifulSoup(r.text, 'html.parser')
with open('./test.csv', 'w+') as f:
    for i in soup.find_all('a'):
        print(i.string, end=': ')
        print(i.attrs.get('href'))
        f.write(str(i.string) + ',' + str(i.attrs.get('href')) + '\n')
```

2.	利用numpy随机函数生成一个4行5列数值是（100,200）随机整数的矩阵。对矩阵中的数分别求平方根，平方，取整；使用numpy的统计函数：Sum,mean ,average,std,var，min,max等函数。

```python
import numpy as np
a = np.random.randint(100, 201, 20).reshape(4, 5) #生成4行5列数值是(100,200)随机整数矩阵

print('初始矩阵：\n',a)    # 输出矩阵
print('平方根：\n',np.sqrt(a))   # 平方根
print('平方：\n',np.square(a))  # 平方
print('向上取整：\n',np.ceil(a))  # 向上取整
print('向下取整：\n',np.floor(a))  # 向下取整

print('求和：\n',a.sum())  # 求和
print('平均值：\n',a.mean())  # 平均值
print('平均值：\n',np.average(a))  # 平均值
print('方差：\n',a.std())  # 方差
print('标准差：\n',a.var())  # 标准差
print('最小值：\n',a.min())  # 最小值
print('最大值：\n',a.max())  # 最大值
```

3.	在同一坐标轴中绘制下列函数曲线:Y=2X2(红色虚线), Y=3X3(蓝色虚点画线), Y=4X4(黄色虚破折线)并分别在X轴和Y轴标上中文标签(“X轴. Y轴“)，并加上标题“函数曲线”

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

x = np.linspace(-2, 2)
y1 = 2*np.power(x, 2)
y2 = 3*np.power(x, 3)
y3 = 4*np.power(x, 4)
matplotlib.rcParams['font.family'] = 'KaiTi'  # 显示中文字符
matplotlib.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
plt.figure()
plt.title('函数曲线', fontsize=25)  # 设置标题
plt.xlabel('X轴', fontsize=15)  # 设置x轴标签
plt.ylabel('Y轴', fontsize=15)  # 设置y轴标签
plt.plot(x, y1, 'r:', label='$y=2x^2$')
plt.plot(x, y2, 'b-.', label='$y=3x^3$')
plt.plot(x, y3, 'y--', label='$y=4x^4$')
plt.legend(loc=0)  # 放置图注位置为中间
plt.show()
```

4.编码实现下列三个函数图形：
 



```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 1000)
y = np.cos(2*np.pi*x)
matplotlib.rcParams['font.family'] = 'KaiTi'  # 显示中文字符
matplotlib.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
plt.grid()  # 设置网格线
plt.xlim(-1, 6)  # 设置x轴取值范围
plt.ylim(-2.0, 2.0)  # 设置y轴取值范围
plt.plot(x, y, 'r--')
plt.title('正弦波实例$y=cos(2Πx)$', fontsize=25)  # 设置图表标题
plt.xlabel('横轴：时间', fontsize=15, color='red')  # 设置x轴标签
plt.ylabel('纵轴:振幅', fontsize=15, color='black')  # 设置y轴标签
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 6.27, 200)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x*x)
for i in [[221, y1, 'r'], [222, y2, 'b--'], [212, y3, 'g--']]:
    plt.subplot(i[0])
    plt.plot(x, i[1], i[2])
    plt.xticks(np.arange(0, 8))
    plt.yticks(np.arange(-1.0, 1.5, 0.5))
plt.show()
```
